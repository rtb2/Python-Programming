```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
data = pd.read_csv("A:\\Data Science\\Super Data Science\\Python Programming\\Section6-Homework-Data.csv", encoding='latin1')
```

```python
Gen_data = data[(data.Genre =='action')|(data.Genre =='comedy')|(data.Genre =='adventure')|(data.Genre =='animation')|\
                (data.Genre =='drama')]
```

```python
Gen_data2= Gen_data[(Gen_data.Studio=='Buena Vista Studios')|(Gen_data.Studio=='Sony')|\
                   (Gen_data.Studio=='Universal')|(Gen_data.Studio=='WB')|\
                   (Gen_data.Studio=='Paramount Pictures')|(Gen_data.Studio=='Fox')]
```

```python
Gen_data2.columns
```

```python
# Define the style
sns.set(style="darkgrid", palette="muted", color_codes=True)

# Plot the boxsplots
ax = sns.boxplot(data=Gen_data2, x='Genre', y='Gross % US', orient='v', color='lightgray', showfliers=False)
plt.setp(ax.artists, alpha=0.5)

# Add in points to show each observation
sns.stripplot(x='Genre', y='Gross % US', data=Gen_data2, jitter=True, size=6, linewidth=0, hue = 'Studio', alpha=0.7)

ax.axes.set_title('Domestic Gross % by Genre',fontsize=30)
ax.set_xlabel('Genre',fontsize=20)
ax.set_ylabel('Gross % US',fontsize=20)

# Define where to place the legend
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
